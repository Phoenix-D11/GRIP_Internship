### Using one hot encoding, we encode the categorical data (Species)

```python
species = LabelEncoder()
```

```python
df_iris["Species"] = species.fit_transform(df_iris["Species"])
```

```python
df_iris.sample(10)
```

### Building the model

```python
X = df_iris.drop("Species", axis = "columns")
y = df_iris.Species
```

```python
from sklearn.tree import DecisionTreeClassifier
iris_tree=DecisionTreeClassifier()
iris_tree.fit(X,y)
```

```python
features = ["SepalLengthCm", "SepalWidthCm",  "PetalLengthCm", "PetalWidthCm"]
iris=['Iris-Setosa','Iris-Versicolor','Iris-virginica']
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (6,6), dpi = 450)

graph = tree.plot_tree(iris_tree, feature_names = features, class_names = iris, filled = True);
```
